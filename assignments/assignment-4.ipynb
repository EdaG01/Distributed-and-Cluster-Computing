### Assignment 4 (30 points)

In this assignment, each team will extend and redesign the cluster created in Assignment 3 to include shared file systems. 

Part I: Nodes arrangement (5 points)

You are to modify the cluster's design from 4 nodes to  6 nodes. The nodes should be arranged in the following manner:

- 1 head node: has public floating IP address (accessible from outside) and fixed local IP address (subnet flat-lan-1-net)

- 3 compute nodes: has fixed local IP address (subnet flat-lan-1)

- 2 storage nodes: has fixed local IP address (subnet flat-lan-1)

All nodes should have passwordless SSH access to one another

 

Part II: Shared file system (10 points)

You are to set up two shared file systems, one networked file system (NFS), and one parallel file systems (OrangeFS)

- The NFS is originated from the head node, and supports two shared directories: /home and /software. These two directories are to be mounted on the head node and compute nodes.

- The OrangeFS server is to be created on the 2 storage nodes, and forms the directory /scratch. This directory is to be mounted on the head node and compute nodes. 

 

You should also think ahead at this stage and make sure the following information is included:

- Automounting of shared directories

- Inclusion of PATH on all compute nodes regarding software available through /software
